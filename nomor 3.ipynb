```python
def sum_product(a,b):
    s = 0
    for i in range(n):
        s += a[i]*b[i]
    return s

def linear_regression(x,y):
    yreg = []
    for i in x:
        yreg.append(((sum(y)*sum_product(x,x)-(sum(x)*sum_product(x,y)))/((n*sum_product(x,x))-(sum(x)*sum(x)))) + (((n*sum_product(x,y))-(sum(x)*sum(y)))/((n*sum_product(x,x))-(sum(x)*sum(x)))) * i)
    return yreg
```

```python
import numpy as np

def q(j,x1,x,y):
    q1 = y[j+1]*(x1-x[j]) / (x[j+1]-x[j])
    q2 = y[j]*(x1-x[j+1]) / (x[j]-x[j+1])
    qjx = q1+q2
    return qjx

import math
def p(x1,x,y):
    n = min(len(x),len(y))
    h = x[1]-x[0]
    j = math.floor((x1-x[0])/h)
    if j >= n - 1: j = n - 2
    px = q(j,x1,x,y)
    return px

def linear_interpol(x1, x, y):
    y = [p(i,x,y) for i in x1]
    return y
```

```python
def q2(j,x2,x,y):
    dumm = 1
    for k in range(n):
        if k != j:
            dumm *= (x2-x[k]) / (x[j] - x[k])
    qjx = y[j] * dumm
    return qjx

def p2(x2, x, y):
    px = 0
    for j in range(n):
        px += (q2(j,x2,x,y))
    return px

def lagrange_interpol(x2,x,y):
    y = [p2(i,x,y) for i in x2]
    return y
```
